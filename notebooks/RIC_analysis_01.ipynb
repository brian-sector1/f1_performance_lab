```python
import fastf1
session = fastf1.get_session(2018, "Monaco", 'Q')
# the race that max verstappen destroyed his car in practice
session.load()
session.results
session.results.iloc[10:20].loc[:, ['Abbreviation']]
```

```python
session.laps.iloc[30:35]
```

```python
fastest_lap = session.laps.pick_fastest()
print(fastest_lap['LapTime'])
print(fastest_lap['Driver'])
```

```python
from matplotlib import pyplot as plt
import fastf1.plotting
fastf1.plotting.setup_mpl(misc_mpl_mods=False, color_scheme='fastf1')
ricciardo = fastest_lap
ric_car_data = ricciardo.get_car_data();
t = ric_car_data['Time'];
vCar = ric_car_data['Speed'];

fig, ax = plt.subplots()
ax.plot(t,vCar, label='Fastest Lap')
ax.set_xlabel('Time')
ax.set_ylabel('Speed [Km/h]')
ax.set_title('Ricciardo:')
ax.legend()
plt.show()
```
