```python
import fastf1
import fastf1.plotting
import seaborn as sns
from matplotlib import pyplot as plt

fastf1.plotting.setup_mpl(mpl_timedelta_support=True, misc_mpl_mods=False, color_scheme='fastf1')

session1 = fastf1.get_session(2019, "Monaco", 'R')
session1.load()

driver_laps = session1.laps.pick_drivers("VER").pick_quicklaps().reset_index()
```

```python
fig, ax = plt.subplots(figsize = (8, 8))

sns.scatterplot(data=driver_laps,
                x="LapNumber",
                y="LapTime",
                ax=ax,
                hue="Compound",
                palette=fastf1.plotting.get_compound_mapping(session=session1),
                s=80,
                linewidth=0,
                legend='auto')

```
